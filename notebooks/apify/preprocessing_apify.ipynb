```python
import pandas as pd
```

# Xử lý `post_khongbanlaptop_apify`

```python
df_post_apify = pd.read_csv('../../data/apify/posts_apify.csv')
df_post_apify
```

```python
feature_columns2 = ['postId', 'timestamp', 'topLevelUrl', 'text', 'likes', 'comments', 'shares']
```

```python
df_post_apify1 = df_post_apify[feature_columns2].copy()
df_post_apify1['timestamp'] = pd.to_datetime(df_post_apify1['timestamp'], unit='s', utc=True)

df_post_apify1.rename(columns={
    'timestamp': 'postDate',
    'topLevelUrl': 'url'
}, inplace=True)

df_post_apify1
```

```python
post_id = [1147451870760756, 1147447477427862, 1147436614095615, 
            1147413360764607, 1146548730851070, 1146531914186085,
            1146529790852964, 1146504504188826, 1145328240973119]
            
df_filtered_posts = df_post_apify1[df_post_apify1['postId'].isin(post_id)].copy()
df_filtered_posts
```

```python
df_filtered_posts.isna().sum()
```

```python
df_filtered_posts['comments'] = df_filtered_posts['comments'].fillna(0)
df_filtered_posts.loc[df_filtered_posts['text'].isna(), 'text'] = (
    "Thế Giới Di Động (thegioididong.com) updated their cover photo."
)

df_filtered_posts['total_engagement'] = (
    df_filtered_posts['likes'] +
    df_filtered_posts['shares'] +
    df_filtered_posts['comments']
)

df_filtered_posts
```

```python
df_filtered_posts.to_csv('../../data/apify/posts_apify_processed.csv', index=False)
```

# Xử lý `comment_data_apify`

```python
df_comment_apify = pd.read_csv('../../data/apify/comments_apify.csv')
df_comment_apify
```

```python
feature_columns1 = ['date', 'facebookId', 'profileId', 'text', 'likesCount']
```

```python
df_comment_apify1 = df_comment_apify[feature_columns1].copy()
df_comment_apify1.loc[:, 'date'] = pd.to_datetime(df_comment_apify1['date'], errors='coerce')
df_comment_apify1.rename(columns={
    'date': 'cmtDate',
    'facebookId': 'postId',
    'profileId': 'userId',
    'text': 'cmtText',
    'likesCount': 'cmtLikes',
    }, inplace=True)
df_comment_apify1
```

```python
df_comment_apify1 = df_comment_apify1.drop_duplicates(subset=['userId', 'postId', 'cmtText'])
df_comment_apify1
```

```python
df_comment_apify1.isnull().sum()
```

```python
df_comment_apify1 = df_comment_apify1.dropna()
df_comment_apify1
```

```python
df_comment_apify1.to_csv('../../data/apify/comments_apify_processed.csv', index=False)
```
